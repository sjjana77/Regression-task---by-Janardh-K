## Heatmap

```python
plt.figure(figsize=(11,7))
sns.heatmap(data.corr(),annot=True)
plt.tight_layout()
```

## Countplot for Region

```python
plt.figure(figsize=(9,7))
sns.countplot(x="Region", data=data, palette='plasma')
plt.title("Region Count", fontsize=15)
plt.xlabel("Region",fontsize=15)
plt.ylabel("Count of Region", fontsize=15)
plt.tight_layout()
```

## Countplot for Shipping mode

```python

plt.figure(figsize=(8.5,7.5))
sns.countplot('Ship Mode',data=data)
plt.title('Popular shipping modes',size=20)
plt.xlabel('\n Shipping mode',size=10)
plt.ylabel('Number of orders',size=10)
plt.xticks(fontsize=15)
```

## Histogramplot

```python
data.hist(bins=50, figsize=(20,10))
plt.show()
```

## Categoricalplot for Shipmode count

```python
sns.catplot('Ship Mode',data=data,hue='Segment',kind='count',palette='deep',aspect=1.5,height=6)
```

## Piechart

```python
plt.figure(figsize=(10,7))
data['Sub-Category'].value_counts().plot(kind='pie',autopct='%1.1f%%')
plt.show()
```

## Categoricalplot for Discount

```python
plt.figure(figsize=(10,8))
ch=sns.countplot(x='Discount', data=data, order=data['Discount'].value_counts().index,palette="bright")
plt.title("Discount Count", fontsize=15)
plt.xlabel("Discount",fontsize=15)
plt.ylabel("Count of Discount", fontsize=15)
plt.tight_layout()
```

## Categoricalplot for Top 20 State count

```python
plt.figure(figsize=(12,10))
sns.countplot(x=data['State'], order=(data['State'].value_counts().head(20)).index)
plt.xticks(rotation=90)
plt.title("Top 20 State Count")
plt.tight_layout()
```

## Categoricalplot for State count

```python
sns.catplot('State', data=data, kind='count', palette='muted',height=8,aspect=2)
plt.xticks(rotation=90)
```

## Top 10 profitable Cities

```python
df_top10_cities = data.groupby('City')['Profit'].sum().reset_index().sort_values(by='Profit',ascending=False)
df_top10_cities = df_top10_cities.head(10)
df_top10_cities
```

## Categorialplot for top 10 profitable Cities

```python
sns.catplot('City','Profit',data=df_top10_cities,kind='bar',aspect=2,height=8,palette='colorblind')
plt.title('Top 10 cities with profit inflow',size=25)
plt.xticks(size=15)
plt.yticks(size=15)
plt.ylabel('Cumulative profit',size=18)
plt.xlabel('City',size=18)
```
